```python
def imdbid_mdbid(imdbid): 
    url="https://api.themoviedb.org/3/find/"+imdbid+"?api_key="+api_key+"&external_source=imdb_id"
    r = requests.get(url)
    json_data=r.json()
    
    mdb_id= str(json_data['movie_results'][0]['id'])
    #this takes in a Imdb-ID and turns it into a a Movie Database ID
    
    url2= "https://api.themoviedb.org/3/movie/"+mdb_id+"?api_key="+api_key
    r2 = requests.get(url2)
    json_data2=r2.json()
    
    return json_data2

imdbid_mdbid('tt0111161')
```

```python
listings = soup4.find_all('div', class_='wlb_ribbon')   # Created a list of id#s for the bottom 100

bottom_100_ids=[]
for listing in listings:
     bottom_100_ids.append(listing.get("data-tconst"))

bottom_100_ids
```

```python
import time

worst_movies= []
for movie_id in bottom_100_ids:
    #print(i)
    worst_movies.append(imdbid_mdbid(movie_id))
    time.sleep(1.5)
```
